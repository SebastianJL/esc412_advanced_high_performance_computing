```python
import numpy as np
import matplotlib.pyplot as plt
mass_grid_modulo = np.genfromtxt('mass_grid_modulo_2d.csv', delimiter=',')
mass_grid_margins = np.genfromtxt('mass_grid_with_margins_2d.csv', delimiter=',')

diff = mass_grid_modulo - mass_grid_margins

datas = [np.log(mass_grid_modulo + 1), 
         np.log(mass_grid_margins + 1),
        diff
        ]

print(mass_grid_margins.max())
print(diff[diff < -1])
print(diff[diff > 0])

for data in datas:
    print(f'{data.shape = }')

fig, axes = plt.subplots(1, 3, figsize=(15, 15))
titles = ['modulo', 'with margins', 'difference']
cmaps = [None, None, 'PiYG']
for ax, title, data, cmap in zip(axes, titles, datas, cmaps):

    pos = ax.imshow(data, cmap=cmap)
    ax.title.set_text(title)
    fig.colorbar(pos, ax=ax, shrink=0.5)

```
